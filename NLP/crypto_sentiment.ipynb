## 1. Sentiment Analysis

Use the [newsapi](https://newsapi.org/) to pull the latest news articles for Bitcoin and Ethereum and create a DataFrame of sentiment scores for each coin.

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from dotenv import load_dotenv
import nltk as nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

%matplotlib inline
```

```python
from newsapi import NewsApiClient
```

```python
load_dotenv()
```

```python
# Read your api key environment variable
api_key = os.getenv('newsapi_key')
type(api_key)
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch the Bitcoin news articles
bitcoin_headlines = newsapi.get_everything(q = 'bitcoin', language = 'en')
bitcoin_headlines
```

```python
# Fetch the Ethereum news articles
ethereum_headlines = newsapi.get_everything(q = 'ethereum', language = 'en')
```

```python
nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()
```

```python
# Create the Bitcoin sentiment scores DataFrame
bitcoin_sentiments = []
for article in bitcoin_headlines['articles']:
    try:
        text = article['content']
        date = article['publishedAt'][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment['compound']
        pos = sentiment['pos']
        neu = sentiment['neu']
        neg = sentiment['neg']
        
        bitcoin_sentiments.append({
            'text': text,
            'date': date,
            'compound': compound,
            'positive': pos,
            'negative': neg,
            'neutral': neu
            
        })
        
    except AttributeError:
        pass
    
bitcoin_df_original = pd.DataFrame(bitcoin_sentiments)
cols = ['date', "text", "compound", "positive", "negative", "neutral"]
bitcoin_df = bitcoin_df_original[cols]
bitcoin_df
```

```python
# Create the Ethereum sentiment scores DataFrame
ethereum_sentiments = []
for article in ethereum_headlines['articles']:
    try:
        text = article['content']
        date = article['publishedAt'][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment['compound']
        pos = sentiment['pos']
        neu = sentiment['neu']
        neg = sentiment['neg']
        
        ethereum_sentiments.append({
            'text': text,
            'date': date,
            'compound': compound,
            'positive': pos,
            'negative': neg,
            'neutral': neu
            
        })
        
    except AttributeError:
        pass
    
ethereum_df_original = pd.DataFrame(ethereum_sentiments)
cols = ['date', "text", "compound", "positive", "negative", "neutral"]
ethereum_df = ethereum_df_original[cols]
ethereum_df
```

```python
# Describe the Bitcoin Sentiment
bitcoin_df.describe()
```

```python
# Describe the Ethereum Sentiment
ethereum_df.describe()
```

###   Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word.
2. Remove Punctuation.
3. Remove Stopwords.

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
import nltk
nltk.download('wordnet')
```

```python
# Instantiate the lemmatizer
lemmatizer = WordNetLemmatizer()

# Create a list of stopwords
nltk.download('stopwords')

# Expand the default stopwords list if necessary
sw_adds = {''}
```

```python
import string
```

```python
# Complete the tokenizer function
def tokenizer(text, sw_adds):
    """Tokenizes text."""
    
    # Remove the punctuation from text
    for punctuation in string.punctuation:
        text = text.replace(f'{punctuation}', '')
   
    # Create a tokenized list of the words
    words = word_tokenize(text)
    
    # Lemmatize words into root words
    # Convert the words to lowercase
    # Remove the stop words
    sw = set(stopwords.words('english'))
    words = [lemmatizer.lemmatize(word.lower()) for word in words if word not in sw.union(sw_adds)]
    
    return words
```

```python
# Create a new tokens column for Bitcoin
bitcoin_df['tokens'] = [tokenizer(text, '') for text in bitcoin_df['text']]
bitcoin_df
```

```python
# Create a new tokens column for Ethereum
ethereum_df['tokens'] = [tokenizer(text, '') for text in ethereum_df['text']]
ethereum_df
```

```python
# Function token_count generates the top 10 words for a given coin
def token_count(df, n = 3):
    """Returns the top N tokens from the frequency count"""

    most_common = []
    for token_list in df['tokens']:
        most_common.append(Counter(token_list).most_common(n))
        
    token_counts = {}
    for list in most_common:
        for tuple in list:
            if tuple[0] not in token_counts:
                token_counts[tuple[0]] = tuple[1]
            if tuple[0] in token_counts:
                token_counts[tuple[0]] += tuple[1]

    sorted_token_counts = dict(sorted(token_counts.items(), key = lambda x: x[1], reverse = True))

    return [(k, v) for k, v in sorted_token_counts.items()]
```

```python
# Function token_count generates the top 10 words for a given coin
def token_count_2(df, n = 3):
    """Returns the top N tokens from the frequency count"""

    most_common = []
    for token_list in df['tokens']:
        most_common.append(Counter(token_list).most_common(n))
        
    token_counts = {}
    for item in most_common:
        for tup in item:
            if tup[0] not in token_counts:
                token_counts[tup[0]] = tup[1]
            if tup[0] in token_counts:
                token_counts[tup[0]] += tup[1]

    return list(token_counts.items())
```

```python
# Use token_count to get the top 10 words for Bitcoin
token_count(bitcoin_df, 10)
```
